## Objectif de cet acte

**Probl√®me :** Les algorithmes ne comprennent que les nombres. Comment leur faire "lire" du texte ?

**Solution :** Transformer chaque texte en un **vecteur de nombres**.

| Texte | ‚Üí | Vecteur |
|-------|---|--------|
| "J'adore Python" | ‚Üí | [1, 0, 0, 1, 1] |

C'est le **Bag of Words** (sac de mots) : la premi√®re technique de repr√©sentation de texte.

---

# PARTIE 1 : L'intuition

---

## 1.1 üéí La m√©taphore du sac

Imaginez que vous avez une phrase √©crite sur des cartes, un mot par carte.

Vous mettez toutes les cartes dans un sac, vous secouez... et vous comptez combien de fois chaque mot appara√Æt.

**Vous avez perdu l'ordre des mots, mais vous savez lesquels sont pr√©sents et combien de fois.**

C'est exactement ce que fait le Bag of Words !

```python
# Illustrons avec un exemple
phrase = "Les enfants adorent Python et les enfants codent"

print("üìù PHRASE ORIGINALE")
print(f"   '{phrase}'")
print(f"   -> L'ordre compte : 'Les' est au d√©but, 'codent' √† la fin")

# Mettons les mots dans le "sac" (on compte)
from collections import Counter
mots = phrase.lower().split()
sac = Counter(mots)

print("\nüéí DANS LE SAC (Bag of Words)")
print(f"   {dict(sac)}")
print(f"   -> L'ordre est perdu, mais on sait que 'les' appara√Æt 2 fois, 'enfants' 2 fois...")
```

## 1.2 La cons√©quence importante

Puisqu'on perd l'ordre, certaines phrases **diff√©rentes** deviennent **identiques** !

```python
# Deux phrases avec des sens TR√àS diff√©rents
phrase1 = "Le chat mange la souris"
phrase2 = "La souris mange le chat"

print("üß† POUR UN HUMAIN")
print(f"   Phrase 1 : '{phrase1}'")
print(f"   Phrase 2 : '{phrase2}'")
print(f"   ‚Üí Ces phrases ont des sens OPPOS√âS !")

# Dans le sac...
sac1 = Counter(phrase1.lower().split())
sac2 = Counter(phrase2.lower().split())

print("\nüéí POUR LE BAG OF WORDS")
print(f"   Sac 1 : {dict(sac1)}")
print(f"   Sac 2 : {dict(sac2)}")
print(f"   ‚Üí Identiques ? {sac1 == sac2}")
print("\n   ‚ö†Ô∏è Le Bag of Words ne distingue PAS ces deux phrases !")
```

## 1.3 Pourquoi utiliser une technique si limit√©e ?

Parce que malgr√© cette limite, le Bag of Words est :

1. **Simple** √† comprendre et √† impl√©menter
2. **Rapide** √† calculer, m√™me sur des millions de documents
3. **Efficace** pour beaucoup de t√¢ches (recherche, classification, etc.)

Et surtout : c'est la **premi√®re √©tape** pour comprendre les techniques plus avanc√©es !

# PARTIE 2 : Construction pas √† pas

---

Voyons comment transformer concr√®tement un texte en vecteur.

## 2.1 Notre corpus d'exemple

Un **corpus** est simplement une collection de documents (textes).

Prenons des retours d'√©l√®ves du programme SUCCEED :

```python
# Notre mini-corpus : 3 retours d'√©l√®ves
corpus = [
    "j'adore Python",
    "Python est facile",
    "j'adore apprendre"
]

print("üìö CORPUS : 3 documents")
print("=" * 40)
for i, doc in enumerate(corpus):
    print(f"   Document {i} : '{doc}'")
```

## 2.2 √âtape 1 : Construire le vocabulaire

Le **vocabulaire** est la liste de tous les mots **uniques** pr√©sents dans le corpus.

Chaque mot re√ßoit un **index** (une position fixe).

```python
# √âtape 1 : Collecter tous les mots uniques
def construire_vocabulaire(corpus):
    """
    Parcourt tous les documents et collecte les mots uniques.
    Retourne une liste tri√©e (pour avoir un ordre coh√©rent).
    """
    tous_les_mots = set()  # Un set ne garde que les valeurs uniques
    
    for document in corpus:
        mots = document.lower().split()  # D√©couper en mots
        tous_les_mots.update(mots)       # Ajouter au set
    
    return sorted(tous_les_mots)  # Trier pour coh√©rence

# Construisons le vocabulaire
vocabulaire = construire_vocabulaire(corpus)

print("üìñ √âTAPE 1 : VOCABULAIRE")
print("=" * 40)
print(f"   Mots uniques trouv√©s : {vocabulaire}")
print(f"   Taille du vocabulaire : {len(vocabulaire)} mots")
```

```python
# Cr√©er le mapping : mot ‚Üí index (position dans le vecteur)
mot_vers_index = {mot: index for index, mot in enumerate(vocabulaire)}

print(" MAPPING : Chaque mot a une position fixe")
print("=" * 40)
for mot, index in mot_vers_index.items():
    print(f"   '{mot}' ‚Üí position {index}")

print("\n   üí° Cette position sera la m√™me pour TOUS les documents !")
```

## 2.3 √âtape 2 : Vectoriser un document

Pour chaque document, on cr√©e un vecteur :
- **Taille** = nombre de mots dans le vocabulaire
- **Valeur √† chaque position** = nombre de fois que le mot correspondant appara√Æt

### Comment √ßa marche ?

1. On cr√©e un vecteur rempli de z√©ros
2. On parcourt le texte mot par mot
3. Pour chaque mot, on incr√©mente la case correspondante

```python
def vectoriser(texte, mot_vers_index):
    """
    Transforme un texte en vecteur de comptage.
    
    Args:
        texte: Le document √† vectoriser
        mot_vers_index: Le dictionnaire {mot: position}
    
    Returns:
        Un vecteur (liste) o√π chaque position = count du mot correspondant
    """
    # 1. Cr√©er un vecteur de z√©ros
    taille_vocabulaire = len(mot_vers_index)
    vecteur = [0] * taille_vocabulaire
    
    # 2. Parcourir le texte mot par mot
    for mot in texte.lower().split():
        # 3. Si le mot est dans notre vocabulaire, incr√©menter
        if mot in mot_vers_index:
            position = mot_vers_index[mot]
            vecteur[position] += 1
    
    return vecteur

print("Fonction de vectorisation cr√©√©e !")
```

```python
# Vectorisons le premier document pas √† pas
doc = corpus[0]  # "j'adore Python"

print(f"VECTORISATION DE : '{doc}'")
print("=" * 50)
print(f"\n   Vocabulaire : {vocabulaire}")
print(f"   Index :        {list(range(len(vocabulaire)))}")

print("\n   Parcours du texte :")
vecteur = [0] * len(vocabulaire)

for mot in doc.lower().split():
    if mot in mot_vers_index:
        pos = mot_vers_index[mot]
        vecteur[pos] += 1
        print(f"      ‚Ä¢ Mot '{mot}' trouv√© ‚Üí position {pos} ‚Üí vecteur = {vecteur}")

print(f"\n   Vecteur final : {vecteur}")
```

```python
# Vectorisons tous les documents
print("VECTORISATION DE TOUT LE CORPUS")
print("=" * 50)
print(f"   Vocabulaire : {vocabulaire}")
print()

for i, doc in enumerate(corpus):
    vecteur = vectoriser(doc, mot_vers_index)
    print(f"   Doc {i} : '{doc}'")
    print(f"         -> {vecteur}")
    print()
```

## 2.4 Visualisation : La matrice document-terme

Quand on vectorise tous les documents, on obtient une **matrice** :
- Chaque **ligne** = un document
- Chaque **colonne** = un mot du vocabulaire
- Chaque **cellule** = combien de fois le mot appara√Æt dans le document

```python
import pandas as pd

# Cr√©er la matrice
matrice = [vectoriser(doc, mot_vers_index) for doc in corpus]

# Afficher avec pandas
df = pd.DataFrame(
    matrice,
    columns=vocabulaire,
    index=[f"Doc {i}: '{corpus[i]}'" for i in range(len(corpus))]
)

print("üìã MATRICE DOCUMENT-TERME")
print("=" * 60)
print(df)
print("\n Lecture : Doc 0 contient 1 fois 'adore', 0 fois 'apprendre', etc.")
```

# PARTIE 3 : La formule math√©matique

---

## 3.1 Formalisation

Pour un document $d$ et un vocabulaire $V = \{w_1, w_2, ..., w_n\}$ :

$$\Large \text{BoW}(d) = [\text{count}(w_1, d), \text{count}(w_2, d), ..., \text{count}(w_n, d)]$$

O√π :
- $\text{count}(w_i, d)$ = nombre de fois que le mot $w_i$ appara√Æt dans le document $d$
- $n$ = taille du vocabulaire

```python
# Exemple avec un document qui r√©p√®te des mots
doc_exemple = "Python Python Python est facile tr√®s facile"

# Nouveau vocabulaire pour cet exemple
vocab_exemple = construire_vocabulaire([doc_exemple])
mapping_exemple = {mot: i for i, mot in enumerate(vocab_exemple)}

print(" APPLICATION DE LA FORMULE")
print("=" * 50)
print(f"   Document : '{doc_exemple}'")
print(f"   Vocabulaire V = {vocab_exemple}")
print()

# Calculer count pour chaque mot
print("   Calcul de count(wi, d) pour chaque mot :")
mots_doc = doc_exemple.lower().split()
for mot in vocab_exemple:
    count = mots_doc.count(mot)
    print(f"      count('{mot}', d) = {count}")

vecteur_exemple = vectoriser(doc_exemple, mapping_exemple)
print(f"\n   BoW(d) = {vecteur_exemple}")
```

# PARTIE 4 : Avec scikit-learn

---

En pratique, on utilise des biblioth√®ques comme **scikit-learn** qui font tout √ßa automatiquement.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Notre corpus
corpus = [
    "j'adore Python",
    "Python est facile",
    "j'adore apprendre"
]

# Cr√©er le vectoriseur
vectorizer = CountVectorizer()

# fit_transform = construire le vocabulaire ET vectoriser
X = vectorizer.fit_transform(corpus)

print("AVEC SCIKIT-LEARN")
print("=" * 50)
print(f"   Vocabulaire : {vectorizer.get_feature_names_out()}")
print(f"   Matrice shape : {X.shape} (3 documents √ó {X.shape[1]} mots)")
```

```python
# Afficher la matrice
df_sklearn = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out(),
    index=[f"Doc {i}" for i in range(len(corpus))]
)

print("üìã MATRICE DOCUMENT-TERME (scikit-learn)")
print(df_sklearn)
print("\n M√™me r√©sultat qu'√† la main, en 2 lignes de code !")
```

```python
# Vectoriser un nouveau document
nouveau_doc = ["Python est super"]

# ATTENTION : on utilise transform() (pas fit_transform)
# car le vocabulaire est d√©j√† construit
X_nouveau = vectorizer.transform(nouveau_doc)

print("VECTORISER UN NOUVEAU DOCUMENT")
print("=" * 50)
print(f"   Document : '{nouveau_doc[0]}'")
print(f"   Vocabulaire existant : {list(vectorizer.get_feature_names_out())}")
print(f"   Vecteur : {X_nouveau.toarray()[0]}")
print("\n   'super' n'est pas dans le vocabulaire ‚Üí ignor√© !")
```

# PARTIE 6 : Une question importante ‚Äî Le poids des mots

---

## 6.1 ‚öñÔ∏è Qu'est-ce qu'un "poids" ?

Vous entendrez souvent parler de **poids** en machine learning. C'est un concept simple :

> Un **poids** est un nombre qui indique **l'importance** d'un √©l√©ment.

### Analogie : La balance de courses

Imaginez que vous faites vos courses. Tous les produits n'ont pas la m√™me importance pour vous :
- Le pain ‚Üí **tr√®s important** (poids √©lev√©)
- Les chips ‚Üí **peu important** (poids faible)

Si on vous demande "Vos courses sont-elles r√©ussies ?", vous allez **pond√©rer** :
- Avoir le pain compte beaucoup
- Avoir les chips compte peu

```python
# Illustration : Poids dans la vie quotidienne
print("‚öñÔ∏è ANALOGIE : LA BALANCE DE COURSES")
print("=" * 50)

# Vos courses
courses = {"pain": 1, "lait": 1, "chips": 0, "fruits": 1}  # 1 = achet√©, 0 = oubli√©

# Sans poids : tout compte pareil
score_simple = sum(courses.values()) / len(courses)
print(f"\n   Sans poids : {score_simple:.0%} de r√©ussite")
print(f"   -> 'chips' compte autant que 'pain'")

# Avec poids : certains produits comptent plus
poids = {"pain": 3, "lait": 2, "chips": 0.5, "fruits": 2}
score_pondere = sum(courses[p] * poids[p] for p in courses) / sum(poids.values())
print(f"\n   Avec poids : {score_pondere:.0%} de r√©ussite")
print(f"   -> 'pain' (poids=3) compte plus que 'chips' (poids=0.5)")
```

## 6.2  Le probl√®me du Bag of Words

Dans notre vecteur Bag of Words, **tous les mots ont le m√™me poids** : 1.

Le mot **"le"** compte autant que le mot **"Python"** !

```python
# D√©monstration du probl√®me
print(" LE PROBL√àME : TOUS LES MOTS ONT LE M√äME POIDS")
print("=" * 50)

doc = "Le cours de Python est le meilleur cours"
vec = CountVectorizer()
X = vec.fit_transform([doc])

print(f"\n   Document : '{doc}'")
print(f"\n   Comptage :")
for mot, count in zip(vec.get_feature_names_out(), X.toarray()[0]):
    importance = "‚≠ê informatif" if mot in ["python", "meilleur"] else "   banal"
    print(f"      '{mot}' : {count} fois  {importance}")

print("\n   -> 'le' (2 fois) semble plus important que 'python' (1 fois) !")
print("   -> C'est absurde : 'python' est bien plus informatif que 'le'")
```

## 6.3 La solution : donner des poids diff√©rents

Et si on donnait **plus de poids** aux mots importants et **moins de poids** aux mots banals ?

| Mot | Poids | Raison |
|-----|-------|--------|
| "le", "de", "est" | Faible | Pr√©sents partout, peu informatifs |
| "Python", "g√©nial" | √âlev√© | Rares et significatifs |

C'est exactement ce que fait **TF-IDF** (Acte I.5) !

```python
print("APER√áU : TF-IDF ")
print("=" * 50)
print("\n   TF-IDF attribue automatiquement des poids :")
print("   ‚Ä¢ Mots fr√©quents PARTOUT ‚Üí poids FAIBLE")
print("   ‚Ä¢ Mots rares mais pr√©sents ‚Üí poids √âLEV√â")
print("\n   Exemple :")
print("      'le'     : pr√©sent dans 90% des docs -> poids ‚âà 0.1")
print("      'Python' : pr√©sent dans 10% des docs -> poids ‚âà 2.3")
print("\n   ‚Üí Rendez-vous √† l'Acte I.5 pour d√©couvrir comment !")
```

# PARTIE 7 : Les limites du Bag of Words

---

## 7.1 ‚ùå Limite 1 : Perte de l'ordre des mots

```python
# Ces deux phrases ont des sens oppos√©s
phrases = [
    "Le formateur aide l'√©l√®ve",
    "L'√©l√®ve aide le formateur"
]

vec = CountVectorizer()
X = vec.fit_transform(phrases)

print(" LIMITE 1 : Perte de l'ordre")
print("=" * 50)
print(f"   Phrase 1 : '{phrases[0]}'")
print(f"   Phrase 2 : '{phrases[1]}'")
print(f"\n   Vecteur 1 : {X.toarray()[0]}")
print(f"   Vecteur 2 : {X.toarray()[1]}")
print(f"\n   Vecteurs identiques ? {(X.toarray()[0] == X.toarray()[1]).all()}")
print("\n   -> Le Bag of Words ne peut pas distinguer ces phrases !")
```

## 7.2 ‚ùå Limite 2 : Les n√©gations sont ignor√©es

```python
# La n√©gation change compl√®tement le sens
phrases = [
    "Ce cours est difficile",
    "Ce cours n'est pas difficile"
]

vec = CountVectorizer()
X = vec.fit_transform(phrases)

print(" LIMITE 2 : N√©gations ignor√©es")
print("=" * 50)
print(f"   Phrase 1 : '{phrases[0]}' (sens n√©gatif)")
print(f"   Phrase 2 : '{phrases[1]}' (sens positif !)")
print(f"\n   Vocabulaire : {list(vec.get_feature_names_out())}")
print(f"\n   -> 'difficile' est pr√©sent dans les DEUX phrases")
print("   -> Le 'pas' n'annule pas le sens de 'difficile'")
```

## 7.3 ‚ùå Limite 3 : Pas de compr√©hension s√©mantique

```python
print(" LIMITE 3 : Pas de s√©mantique")
print("=" * 50)
print("\n   Pour le Bag of Words :")
print("   ‚Ä¢ 'bon' et 'excellent' sont COMPL√àTEMENT diff√©rents")
print("     (alors que ce sont des synonymes)")
print("\n   ‚Ä¢ 'chaud' et 'froid' sont aussi diff√©rents que 'chaud' et 'python'")
print("     (alors que chaud/froid sont des antonymes li√©s)")
print("\n   -> Le BoW ne comprend pas les RELATIONS entre les mots")
```

## 7.4 ‚ùå Limite 4 : Mots nouveaux ignor√©s

```python
# Vocabulaire construit sur un corpus
corpus_train = ["Python est g√©nial", "Java est populaire"]
vec = CountVectorizer()
vec.fit(corpus_train)

# Nouveau document avec un mot inconnu
nouveau_doc = ["JavaScript est fantastique"]
X_nouveau = vec.transform(nouveau_doc)

print(" LIMITE 4 : Mots nouveaux ignor√©s")
print("=" * 50)
print(f"   Vocabulaire appris : {list(vec.get_feature_names_out())}")
print(f"   Nouveau document : '{nouveau_doc[0]}'")
print(f"\n   Vecteur obtenu : {X_nouveau.toarray()[0]}")
print("\n   -> 'JavaScript' et 'fantastique' sont IGNOR√âS !")
```

## 7.5 ‚ùå Limite 5 : Vecteurs creux (Sparsit√©)

Un document moyen contient ~20 mots uniques. Mais le vocabulaire peut contenir 10 000+ mots.

R√©sultat : **plus de 99% des cases du vecteur sont des z√©ros** !

C'est ce qu'on appelle la **sparsit√©** (ou parcimonie).

```python
# Visualisons la sparsit√© avec le corpus SUCCEED
import matplotlib.pyplot as plt
import numpy as np

# Corpus SUCCEED
corpus_succeed = [
    "J'ai ador√© ce cours de Python",
    "Les formateurs sont super",
    "Python est facile √† apprendre",
    "Je n'ai pas compris les exercices",
    "Le cours √©tait trop rapide",
    "J'ai beaucoup appris merci",
    "Les exercices pratiques sont excellents",
    "Je suis perdu avec les boucles",
    "C'√©tait g√©nial et enrichissant",
    "Le contenu √©tait confus et difficile"
]

# Vectoriser (limit√© √† 15 mots pour la lisibilit√©)
vec_sparse = CountVectorizer(max_features=15)
X_sparse = vec_sparse.fit_transform(corpus_succeed)
matrix = X_sparse.toarray()
words = vec_sparse.get_feature_names_out()

# Calculer la sparsit√©
total_cells = matrix.size
zero_cells = np.sum(matrix == 0)
sparsity = (zero_cells / total_cells) * 100

print("LIMITE 5 : Vecteurs creux (Sparsit√©)")
print("=" * 50)
print(f"\n   Matrice : {matrix.shape[0]} documents √ó {matrix.shape[1]} mots")
print(f"   Total de cases : {total_cells}")
print(f"   Cases √† z√©ro : {zero_cells}")
print(f"\n   üî¥ SPARSIT√â : {sparsity:.1f}% de z√©ros !")
```

```python
# Visualisation de la matrice creuse
fig, ax = plt.subplots(figsize=(14, 7))

# Couleurs : rouge fonc√© pour 0, vert pour valeurs > 0
colors = np.where(matrix == 0, 0, 1)
cmap = plt.cm.colors.ListedColormap(['#991b1b', '#15803d'])  # Rouge, Vert

# Afficher la matrice
im = ax.imshow(colors, cmap=cmap, aspect='auto')

# Ajouter les valeurs dans chaque cellule
for i in range(len(corpus_succeed)):
    for j in range(len(words)):
        value = matrix[i, j]
        ax.text(j, i, str(value), ha='center', va='center', 
                color='white', fontsize=11, fontweight='bold')

# Configuration des axes
ax.set_xticks(np.arange(len(words)))
ax.set_yticks(np.arange(len(corpus_succeed)))
ax.set_xticklabels(words, fontsize=10, rotation=45, ha='right')
ax.set_yticklabels([f"Avis {i+1}" for i in range(len(corpus_succeed))], fontsize=10)

# Labels
ax.set_xlabel('Mots du vocabulaire', fontsize=12, fontweight='bold', labelpad=10)
ax.set_ylabel('Documents', fontsize=12, fontweight='bold', labelpad=10)
ax.set_title(f'Matrice Bag of Words ‚Äî Corpus SUCCEED\n{sparsity:.0f}% de cases sont des z√©ros (sparsit√©)', 
             fontsize=14, fontweight='bold', pad=15, color='#10274a')

# Grille
ax.set_xticks(np.arange(len(words)+1)-.5, minor=True)
ax.set_yticks(np.arange(len(corpus_succeed)+1)-.5, minor=True)
ax.grid(which='minor', color='white', linestyle='-', linewidth=2)

# L√©gende
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#991b1b', label='0 (absent)'),
                   Patch(facecolor='#15803d', label='‚â•1 (pr√©sent)')]
ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.12, 1))

plt.tight_layout()
plt.show()

print("\nüí° Observation : La matrice est majoritairement ROUGE (z√©ros) !")
print("   Chaque document n'utilise qu'une petite partie du vocabulaire.")
```

```python
# En situation r√©elle, c'est encore pire !
print("üìà EN SITUATION R√âELLE")
print("=" * 50)

# Simulation
vocab_reel = 10000  # Vocabulaire typique
mots_par_doc = 20   # Mots uniques par document moyen

sparsity_reelle = ((vocab_reel - mots_par_doc) / vocab_reel) * 100

print(f"\n   Vocabulaire typique : {vocab_reel:,} mots")
print(f"   Mots uniques par document : ~{mots_par_doc}")
print(f"\n   üî¥ SPARSIT√â R√âELLE : {sparsity_reelle:.1f}% de z√©ros !")
print("\n   Cons√©quences :")
print("   ‚Ä¢ Beaucoup de m√©moire gaspill√©e")
print("   ‚Ä¢ Calculs sur des millions de z√©ros inutiles")
print("   ‚Ä¢ Solutions : matrices creuses (sparse matrices)")
```

# R√©capitulatif

---

| **BAG OF WORDS** | |
|---|---|
| **Principe** | Transformer un texte en vecteur en comptant les mots. L‚Äôordre des mots est ignor√©. |
| **Formule** | BoW(d) = [count(w‚ÇÅ,d), count(w‚ÇÇ,d), ..., count(w‚Çô,d)] |
| **Poids** | Un poids repr√©sente l‚Äôimportance d‚Äôun mot.  <br> Dans BoW, tous les mots ont le m√™me poids (= 1).  <br> TF-IDF permet d‚Äôattribuer des poids diff√©rents. |
| **‚úÖ Forces** | Simple √† comprendre<br>Rapide √† calculer<br>Fonctionne bien en pratique |
| **‚ùå Limites** | Perd l‚Äôordre des mots<br>Ignore les n√©gations<br>Pas de s√©mantique<br>Mots nouveaux ignor√©s<br>Vecteurs creux (sparsit√©) |
