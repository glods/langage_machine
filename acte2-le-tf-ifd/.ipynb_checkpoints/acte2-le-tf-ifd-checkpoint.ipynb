## Objectif de cet acte

**Problème :** Dans le Bag of Words, tous les mots ont le même poids. Le mot "le" compte autant que "Python".

**Solution :** Donner **plus de poids** aux mots importants et **moins de poids** aux mots banals.

| Mot | Bag of Words | TF-IDF |
|-----|-------------|--------|
| "le" | 1 | 0.05 |
| "Python" | 1 | 2.30 |

C'est le **TF-IDF** : Term Frequency - Inverse Document Frequency.

# PARTIE 1 : L'intuition

---

## 1.1 Le problème du Bag of Words

Rappelons le problème que nous avons identifié dans l'Acte I :

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Un document sur Python
doc = "Le cours de Python est le meilleur cours de programmation"

vec = CountVectorizer()
X = vec.fit_transform([doc])

# Afficher les comptages
df = pd.DataFrame({
    'Mot': vec.get_feature_names_out(),
    'Count': X.toarray()[0]
}).sort_values('Count', ascending=False)

print(" BAG OF WORDS : Comptage simple")
print("=" * 40)
print(df.to_string(index=False))
print("\n----------------------'le' et 'de' apparaissent autant que 'python' !")
print("   Pourtant, 'python' est bien plus informatif...")
```

## 1.2 L'idée clé de TF-IDF

TF-IDF repose sur une intuition simple :

> **Un mot est important s'il est fréquent dans CE document, mais rare dans les AUTRES documents.**

### Deux composantes :

| Composante | Question | Intuition |
|------------|----------|----------|
| **TF** (Term Frequency) | Ce mot apparaît-il souvent dans ce document ? | Plus c'est fréquent, plus c'est important |
| **IDF** (Inverse Document Frequency) | Ce mot apparaît-il dans beaucoup de documents ? | Plus c'est rare, plus c'est important |

```python
print(" ---- L'INTUITION TF-IDF")
print("=" * 50)
print("\n   Imaginons 100 avis d'élèves SUCCEED :")
print("\n   • Le mot 'le' apparaît dans 95 avis sur 100")
print("     → Très commun, peu informatif → Poids FAIBLE")
print("\n   • Le mot 'Python' apparaît dans 10 avis sur 100")
print("     → Plus rare, donc discriminant → Poids ÉLEVÉ")
print("\n   • Le mot 'TensorFlow' apparaît dans 2 avis sur 100")
print("     → Très rare, très spécifique → Poids TRÈS ÉLEVÉ")
```

# PARTIE 2 : Les formules

---

## 2.1 TF — Term Frequency

La fréquence du terme dans le document. C'est simplement le comptage (comme Bag of Words) :

$$\Large \text{TF}(t, d) = \text{count}(t, d)$$

Où :
- $t$ = le terme (mot)
- $d$ = le document
- $\text{count}(t, d)$ = nombre de fois que $t$ apparaît dans $d$

**Variante normalisée** (diviser par le nombre total de mots) :

$$\text{TF}(t, d) = \frac{\text{count}(t, d)}{\text{nombre total de mots dans } d}$$

```python
from collections import Counter

# Exemple de calcul TF
doc = "Python est génial Python est facile Python"
mots = doc.lower().split()

print(" -------------CALCUL DE TF (Term Frequency)")
print("=" * 50)
print(f"   Document : '{doc}'")
print(f"   Nombre total de mots : {len(mots)}")
print("\n   Comptage (TF brut) :")

comptage = Counter(mots)
for mot, count in comptage.most_common():
    tf_normalise = count / len(mots)
    print(f"      TF('{mot}') = {count}  (normalisé : {tf_normalise:.2f})")
```

## 2.2 IDF — Inverse Document Frequency

Mesure la **rareté** d'un mot dans le corpus :

$$\Large \text{IDF}(t) = \log\left(\frac{N}{\text{df}(t)}\right)$$

Où :
- $N$ = nombre total de documents dans le corpus
- $\text{df}(t)$ = nombre de documents contenant le terme $t$

**Interprétation :**
- Si un mot apparaît dans **tous** les documents : $\text{IDF} = \log(N/N) = \log(1) = 0$
- Si un mot apparaît dans **peu** de documents : $\text{IDF}$ est élevé

```python
import numpy as np

# Corpus de 5 documents
corpus = [
    "Le cours de Python est génial",
    "Le formateur est super",
    "Python est facile à apprendre",
    "Le cours était difficile",
    "J'ai adoré le cours de Python"
]

N = len(corpus)  # 5 documents

# Calculer df(t) pour quelques mots
mots_test = ['le', 'python', 'génial', 'tensorflow']

print("----CALCUL DE IDF (Inverse Document Frequency)")
print("=" * 50)
print(f"   Corpus : {N} documents")
print("\n   Calcul de IDF pour chaque mot :")
print("   " + "-" * 45)

for mot in mots_test:
    # Compter dans combien de documents le mot apparaît
    df = sum(1 for doc in corpus if mot.lower() in doc.lower())
    
    if df > 0:
        idf = np.log(N / df)
        print(f"   '{mot}' : présent dans {df}/{N} docs → IDF = log({N}/{df}) = {idf:.2f}")
    else:
        print(f"   '{mot}' : présent dans 0/{N} docs → IDF = ∞ (mot inconnu)")

print("\n  --- Plus le mot est rare, plus l'IDF est élevé !")
```

## 2.3 TF-IDF — La combinaison

On multiplie les deux composantes :

$$\Large \text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)$$

**Résultat :**
- Un mot **fréquent dans ce document** ET **rare dans le corpus** → Score **élevé**
- Un mot **rare dans ce document** OU **commun dans le corpus** → Score **faible**

```python
print("*******FORMULE COMPLÈTE TF-IDF *******")
print("=" * 50)
print("\n   TF-IDF(t, d) = TF(t, d) × IDF(t)")
print("\n   Exemple pour le document 'Python est génial Python' :")
print("   " + "-" * 45)

doc_exemple = "Python est génial Python"
mots_doc = doc_exemple.lower().split()
comptage_doc = Counter(mots_doc)

# Calculer TF-IDF pour chaque mot
for mot in set(mots_doc):
    tf = comptage_doc[mot]
    df = sum(1 for doc in corpus if mot in doc.lower())
    idf = np.log(N / df) if df > 0 else 0
    tfidf = tf * idf
    
    print(f"\n   '{mot}':")
    print(f"      TF = {tf} (apparaît {tf} fois dans ce doc)")
    print(f"      IDF = {idf:.2f} (présent dans {df}/{N} docs)")
    print(f"      TF-IDF = {tf} × {idf:.2f} = {tfidf:.2f}")
```

```python
import matplotlib.pyplot as plt

# Visualiser la fonction IDF
N = 100  # 100 documents
df_values = np.arange(1, 101)  # de 1 à 100 documents
idf_values = np.log(N / df_values)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_values, idf_values, color='#10274a', linewidth=2.5)
ax.fill_between(df_values, idf_values, alpha=0.3, color='#f0a68b')

# Annotations
ax.annotate('Mot rare\n(ex: "TensorFlow")\nIDF élevé', 
            xy=(5, np.log(100/5)), xytext=(20, 3.5),
            fontsize=10, ha='center',
            arrowprops=dict(arrowstyle='->', color='#15803d'),
            color='#15803d', fontweight='bold')

ax.annotate('Mot commun\n(ex: "le")\nIDF faible', 
            xy=(90, np.log(100/90)), xytext=(70, 1.5),
            fontsize=10, ha='center',
            arrowprops=dict(arrowstyle='->', color='#991b1b'),
            color='#991b1b', fontweight='bold')

ax.set_xlabel('Nombre de documents contenant le mot (df)', fontsize=11)
ax.set_ylabel('IDF = log(N/df)', fontsize=11)
ax.set_title('Courbe IDF : Plus un mot est rare, plus son IDF est élevé', 
             fontsize=13, fontweight='bold', color='#10274a')
ax.grid(True, alpha=0.3)
ax.set_xlim(0, 100)
ax.set_ylim(0, 5)

plt.tight_layout()
plt.savefig("courbe.png")

plt.show()
```

# PARTIE 4 : Avec scikit-learn

---

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Corpus SUCCEED
corpus = [
    "J'ai adoré le cours de Python c'était génial",
    "Les formateurs sont super et très pédagogues",
    "Python est facile à apprendre avec cette méthode",
    "Je n'ai pas compris les exercices c'était difficile",
    "Le cours était trop rapide et confus",
    "J'ai beaucoup appris merci aux formateurs",
    "Les exercices pratiques sont excellents",
    "Je suis perdu avec les boucles et les fonctions"
]

# TF-IDF avec scikit-learn
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(corpus)

print(" -------TF-IDF AVEC SCIKIT-LEARN")
print("=" * 50)
print(f"   Corpus : {len(corpus)} documents")
print(f"   Vocabulaire : {len(tfidf.get_feature_names_out())} mots")
print(f"   Matrice shape : {X.shape}")
```

```python
# Comparer Bag of Words vs TF-IDF pour le premier document
doc_idx = 0
doc = corpus[doc_idx]

# Bag of Words
bow = CountVectorizer()
X_bow = bow.fit_transform(corpus)

# Créer un DataFrame comparatif
mots = tfidf.get_feature_names_out()
bow_scores = X_bow[doc_idx].toarray()[0]
tfidf_scores = X[doc_idx].toarray()[0]

# Filtrer les mots présents dans le document
mask = bow_scores > 0
df_compare = pd.DataFrame({
    'Mot': mots[mask],
    'BoW': bow_scores[mask],
    'TF-IDF': tfidf_scores[mask].round(3)
}).sort_values('TF-IDF', ascending=False)

print(f" *******COMPARAISON : Bag of Words vs TF-IDF*******")
print("=" * 50)
print(f"   Document : '{doc}'")
print("\n" + df_compare.to_string(index=False))
print("\n   ********TF-IDF donne plus de poids aux mots discriminants !")
```

# PARTIE 5 : Application — Corpus SUCCEED

---

```python
import random
random.seed(42)

# Générer un corpus synthétique plus grand
templates_positifs = [
    "J'ai adoré le cours de {sujet} c'était {adj}",
    "Le cours de {sujet} était vraiment {adj}",
    "{sujet} est {adj} à apprendre",
    "Les exercices de {sujet} sont {adj}",
    "J'ai beaucoup appris sur {sujet} merci"
]

templates_negatifs = [
    "Je n'ai pas compris le cours de {sujet} c'était {adj}",
    "Le cours de {sujet} était trop {adj}",
    "{sujet} est {adj} à comprendre",
    "Les exercices de {sujet} étaient {adj}",
    "Je suis perdu avec {sujet}"
]

sujets = ['Python', 'JavaScript', 'HTML', 'CSS', 'SQL', 'algorithmes', 'boucles', 'fonctions']
adj_positifs = ['génial', 'super', 'excellent', 'facile', 'clair', 'passionnant', 'enrichissant']
adj_negatifs = ['difficile', 'confus', 'compliqué', 'rapide', 'ennuyeux', 'dur']

# Générer 50 avis positifs et 50 négatifs
corpus_succeed = []
labels = []

for _ in range(50):
    template = random.choice(templates_positifs)
    avis = template.format(sujet=random.choice(sujets), adj=random.choice(adj_positifs))
    corpus_succeed.append(avis)
    labels.append('positif')

for _ in range(50):
    template = random.choice(templates_negatifs)
    avis = template.format(sujet=random.choice(sujets), adj=random.choice(adj_negatifs))
    corpus_succeed.append(avis)
    labels.append('négatif')

print(" ****** CORPUS SUCCEED GÉNÉRÉ ******")
print("=" * 50)
print(f"   Nombre d'avis : {len(corpus_succeed)}")
print(f"   Positifs : {labels.count('positif')}")
print(f"   Négatifs : {labels.count('négatif')}")
print("\n   Exemples :")
for i in [0, 1, 50, 51]:
    print(f"      [{labels[i]}] {corpus_succeed[i]}")
```

```python
# Appliquer TF-IDF
tfidf_succeed = TfidfVectorizer()
X_succeed = tfidf_succeed.fit_transform(corpus_succeed)

# Analyser les IDF (rareté des mots)
mots_succeed = tfidf_succeed.get_feature_names_out()
idf_scores = tfidf_succeed.idf_

# Trier par IDF
df_idf = pd.DataFrame({
    'Mot': mots_succeed,
    'IDF': idf_scores
}).sort_values('IDF', ascending=False)

print(" **** ANALYSE DES IDF *** ")
print("=" * 50)
print("\n   *** Mots les plus RARES (IDF élevé) :")
print(df_idf.head(10).to_string(index=False))
print("\n   **** Mots les plus COMMUNS (IDF faible) :")
print(df_idf.tail(10).to_string(index=False))
```

```python
# Visualiser les IDF
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Top 10 mots rares
top_rare = df_idf.head(10)
axes[0].barh(top_rare['Mot'], top_rare['IDF'], color='#15803d')
axes[0].set_xlabel('IDF (rareté)')
axes[0].set_title('Top 10 mots les plus RARES\n(discriminants)', fontweight='bold', color='#15803d')
axes[0].invert_yaxis()

# Top 10 mots communs
top_commun = df_idf.tail(10).iloc[::-1]
axes[1].barh(top_commun['Mot'], top_commun['IDF'], color='#991b1b')
axes[1].set_xlabel('IDF (rareté)')
axes[1].set_title('Top 10 mots les plus COMMUNS\n(peu informatifs)', fontweight='bold', color='#991b1b')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

## 5.1  Recherche de documents similaires

Une application classique de TF-IDF : trouver les documents les plus similaires à une requête.

```python
from sklearn.metrics.pairwise import cosine_similarity

def rechercher(requete, corpus, tfidf_vectorizer, tfidf_matrix, top_n=3):
    """
    Recherche les documents les plus similaires à une requête.
    """
    # Vectoriser la requête
    requete_vec = tfidf_vectorizer.transform([requete])
    
    # Calculer la similarité cosinus avec tous les documents
    similarites = cosine_similarity(requete_vec, tfidf_matrix)[0]
    
    # Trier par similarité décroissante
    top_indices = similarites.argsort()[::-1][:top_n]
    
    return [(corpus[i], similarites[i]) for i in top_indices]

# Test de recherche
requete = "Python difficile"

print(" RECHERCHE DE DOCUMENTS SIMILAIRES")
print("=" * 50)
print(f"   Requête : '{requete}'")
print("\n  ------ Résultats :")

resultats = rechercher(requete, corpus_succeed, tfidf_succeed, X_succeed, top_n=5)
for i, (doc, score) in enumerate(resultats, 1):
    print(f"\n   {i}. [Score: {score:.3f}]")
    print(f"      {doc}")
```

## 6.1 Différentes formules

Il existe plusieurs variantes de TF-IDF :

| Variante | TF | IDF |
|----------|----|----|  
| **Basique** | count(t,d) | log(N/df) |
| **Normalisé** | count(t,d) / len(d) | log(N/df) |
| **Log** | 1 + log(count(t,d)) | log(N/df) |
| **Scikit-learn** | count(t,d) | log((N+1)/(df+1)) + 1 |

Scikit-learn ajoute des lissages (+1) pour éviter les divisions par zéro et normalise les vecteurs à la fin.

https://www.codeandcortex.fr/tfidf-librairie-scikit-learn/

```python
# Paramètres de TfidfVectorizerhttps://www.codeandcortex.fr/tfidf-librairie-scikit-learn/
print("* ****** PARAMÈTRES DE TfidfVectorizer")
print("=" * 50)
print("""
   TfidfVectorizer(
       norm='l2',           # Normalisation L2 (par défaut)
       use_idf=True,        # Utiliser IDF (sinon = TF seul)
       smooth_idf=True,     # Ajouter 1 au numérateur et dénominateur
       sublinear_tf=False,  # Si True: TF = 1 + log(TF)
       max_df=1.0,          # Ignorer mots présents dans >X% des docs
       min_df=1,            # Ignorer mots présents dans <X docs
   )
""")
```

```python
# Comparaison avec/sans IDF
corpus_test = [
    "Le cours de Python est génial",
    "Le cours de JavaScript est bien",
    "Python et JavaScript sont des langages"
]

# TF seul (sans IDF)
tf_only = TfidfVectorizer(use_idf=False, norm=None)
X_tf = tf_only.fit_transform(corpus_test)

# TF-IDF complet
tfidf_full = TfidfVectorizer(norm=None)
X_tfidf = tfidf_full.fit_transform(corpus_test)

mots = tf_only.get_feature_names_out()

print("******COMPARAISON TF vs TF-IDF")
print("=" * 50)
print(f"   Document : '{corpus_test[0]}'")
print("\n   Mot         TF      TF-IDF")
print("   " + "-" * 30)

for i, mot in enumerate(mots):
    tf_val = X_tf[0, i]
    tfidf_val = X_tfidf[0, i]
    if tf_val > 0:
        marker = "←" if tfidf_val > tf_val else ""
        print(f"   {mot:12} {tf_val:.1f}     {tfidf_val:.2f} {marker}")
```
